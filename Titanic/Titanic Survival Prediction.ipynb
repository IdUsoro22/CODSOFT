<h2> Methodology for Titanic Survival Prediction</h2>

The methodology for analyzing features and building a predictive model in this Titanic Survival Prediction project follows a structured approach to gain insights into the data and understand the relationships between variables that influence survival outcomes.

* **Data Preprocessing and Cleaning:**

    - Checked for missing values and addressed them with various techniques: `Embarked` column missing values were filled with the mode, while `Age` column missing values were filled with the mean using `SimpleImputer`.
    - Created a new feature, `Cabin_Letter`, by extracting the first letter from the `Cabin` column. Missing values in `Cabin_Letter` were imputed based on `Pclass`.
    - Dropped unnecessary columns such as `Cabin`, `PassengerId`, `Name`, and `Ticket` as they provided limited predictive value.

* **Outlier Detection and Treatment:**

    - Identified outliers in numerical columns (`SibSp`, `Parch`, `Fare`) using interquartile range (IQR) thresholds.
    - Replaced detected outliers with calculated threshold values to reduce their impact on model training.

* **Feature Engineering:**

    - Added new features such as `FamilySize` (sum of `SibSp` and `Parch` + 1), `IsAlone` (indicating if a passenger is alone), and `FarePerPerson` (calculated as `Fare` divided by `FamilySize`).
    - Created categorical bins for `Age` using the `pd.cut` function to categorize passengers into age groups (`Child`, `Teen`, `Adult`, `Senior`).

* **Exploratory Data Analysis (EDA):**

    - Visualized distributions of numerical columns using histograms, scatter plots, and boxplots to analyze mean, median, mode, and standard deviation.
    - Conducted correlation analysis on numerical columns and visualized using a heatmap to identify relationships between features and survival.
    - Grouped survival rates by categorical variables (`Pclass`, `Embarked`, `Sex`) to identify patterns in survival likelihood.

* **Model Training and Evaluation:**

    - Split the dataset into training and testing sets with a 67-33 split.
    - Built and evaluated two models: Random Forest Classifier and XGBoost Classifier.
    - For each model, evaluated the classification performance with metrics including accuracy, precision, recall, and F1-score, and visualized confusion matrices to assess model performance.

* **Hyperparameter Tuning:**

    - Defined hyperparameter grids for Random Forest and XGBoost to optimize model performance.
    - Used GridSearchCV with Stratified K-Fold Cross-Validation (5 folds) to search for the best parameters, aiming to maximize accuracy on the validation set.
    - Trained the best models (after tuning) on the full training set and evaluated their final accuracy on the test set, comparing results between the two models.

* **Feature Importance Analysis:**

    - Identified the most important features for survival prediction from the XGBoost model, visualizing feature importance scores to understand which variables had the greatest influence on survival.

```python
# Checking for outliers
def outlier_thresholds(dataframe, col_name, q1=0.05, q3=0.95):
    
    quartile1 = dataframe[col_name].quantile(q1)
    quartile3 = dataframe[col_name].quantile(q3)
    interquantile_range = quartile3 - quartile1
    up_limit = quartile3 + 1.5 * interquantile_range
    low_limit = quartile1 - 1.5 * interquantile_range
    return low_limit, up_limit

def check_outlier(dataframe, col_name):
    
    low_limit, up_limit = outlier_thresholds(dataframe, col_name)
    if dataframe[(dataframe[col_name] > up_limit) | (dataframe[col_name] < low_limit)].any(axis=None):
        return True
    else:
        return False
    
# Check for just numerical cells    
numeric_cols = df.select_dtypes(include=[np.number]).columns


for col in numeric_cols:
    print(col, check_outlier(df, col))
```

```python
# Replacing outliers
def replace_with_thresholds(dataframe, variable):
    low_limit, up_limit = outlier_thresholds(dataframe, variable)
    dataframe.loc[(dataframe[variable] < low_limit), variable] = low_limit
    dataframe.loc[(dataframe[variable] > up_limit), variable] = up_limit

replace_with_thresholds(df, "SibSp")

replace_with_thresholds(df, "Parch")

replace_with_thresholds(df, "Fare")
```

```python
# Recheck for outliers
numeric_cols = df.select_dtypes(include=[np.number]).columns


for col in numeric_cols:
    print(col, check_outlier(df, col))
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

```python
# Initialize model
rf_model = RandomForestClassifier(random_state=42)
```

```python
# Train and evaluate Random Forest
print("Training Random Forest Model...")
rf_model.fit(X_train, y_train)
y_pred_rf = rf_model.predict(X_test)
print("Random Forest Classification Report:")
print(classification_report(y_test, y_pred_rf))
print("Random Forest Accuracy:", accuracy_score(y_test, y_pred_rf))
```

```python
# Confusion matrix for Random Forest
conf_matrix_rf = confusion_matrix(y_test, y_pred_rf)
sns.heatmap(conf_matrix_rf, annot=True, fmt='d', cmap='Blues', xticklabels=['Not Survived', 'Survived'], yticklabels=['Not Survived', 'Survived'])
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.title('Confusion Matrix - Random Forest')
plt.show()
```

# <div style="padding:20px;color:white;margin:0;font-size:35px;text-align:center;display:fill;border-radius:10px;background-color:#c54E58;overflow:hidden; font-family: 'Lucida Console'"><b> HYPER-PARAMETER TUNING</b></div>
<a id="tuning"></a>

```python

# Advanced Feature Engineering
#df['Title'] = df['Name'].str.extract(' ([A-Za-z]+)\.', expand=False)
df['FamilySize'] = df['SibSp'] + df['Parch'] + 1
df['IsAlone'] = (df['FamilySize'] == 1).astype(int)
df['FarePerPerson'] = df['Fare'] / df['FamilySize']
df['AgeGroup'] = pd.cut(df['Age'], bins=[0, 12, 18, 50, 80], labels=['Child', 'Teen', 'Adult', 'Senior'])

# Encode categorical features
df['Sex'] = df['Sex'].map({'male': 0, 'female': 1})
df = pd.get_dummies(df, columns=['Embarked', 'Pclass', 'AgeGroup'], drop_first=True)

#  Select features and target
# X = df[['Sex', 'Age', 'Fare', 'FamilySize', 'IsAlone', 'FarePerPerson', 'Embarked_1', 'Embarked_2', 'Pclass_2', 'Pclass_3']]
# y = df['Survived']

# Define hyperparameter grids for Random Forest and XGBoost
rf_params = {
    'n_estimators': [100, 200, 300],
    'max_depth': [None, 5, 10, 15],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4]
}

xgb_params = {
    'n_estimators': [100, 200, 300],
    'max_depth': [3, 5, 7],
    'learning_rate': [0.01, 0.1, 0.2],
    'subsample': [0.8, 1.0],
    'colsample_bytree': [0.8, 1.0],
    'min_child_weight': [1, 3, 5]
}

# Use Stratified K-Fold Cross-Validation
skf = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)

# Grid Search for Random Forest
rf_grid = GridSearchCV(RandomForestClassifier(random_state=42), rf_params, cv=skf, scoring='accuracy', n_jobs=-1)
rf_grid.fit(X, y)
print("Best Random Forest Parameters:", rf_grid.best_params_)
print("Random Forest Best Cross-Validation Accuracy:", rf_grid.best_score_)

# Grid Search for XGBoost
xgb_grid = GridSearchCV(XGBClassifier(use_label_encoder=False, eval_metric='logloss', random_state=42), xgb_params, cv=skf, scoring='accuracy', n_jobs=-1)
xgb_grid.fit(X, y)
print("Best XGBoost Parameters:", xgb_grid.best_params_)
print("XGBoost Best Cross-Validation Accuracy:", xgb_grid.best_score_)
```

```python
best_rf_model = rf_grid.best_estimator_
best_xgb_model = xgb_grid.best_estimator_

# Train on full training set with best parameters
best_rf_model.fit(X_train, y_train)
best_xgb_model.fit(X_train, y_train)

# Predict on the test set
y_pred_rf = best_rf_model.predict(X_test)
y_pred_xgb = best_xgb_model.predict(X_test)

# Evaluate Random Forest
print("Final Accuracy - Random Forest:", accuracy_score(y_test, y_pred_rf))
print("Classification Report - Random Forest:")
print(classification_report(y_test, y_pred_rf))

# Confusion Matrix for Random Forest
conf_matrix_rf = confusion_matrix(y_test, y_pred_rf)
sns.heatmap(conf_matrix_rf, annot=True, fmt='d', cmap='Blues', xticklabels=['Not Survived', 'Survived'], yticklabels=['Not Survived', 'Survived'])
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.title('Confusion Matrix - Random Forest')
plt.show()

# Evaluate XGBoost
print("Final Accuracy - XGBoost:", accuracy_score(y_test, y_pred_xgb))
print("Classification Report - XGBoost:")
print(classification_report(y_test, y_pred_xgb))

# Confusion Matrix for XGBoost
conf_matrix_xgb = confusion_matrix(y_test, y_pred_xgb)
sns.heatmap(conf_matrix_xgb, annot=True, fmt='d', cmap='Greens', xticklabels=['Not Survived', 'Survived'], yticklabels=['Not Survived', 'Survived'])
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.title('Confusion Matrix - XGBoost')
plt.show()
```
