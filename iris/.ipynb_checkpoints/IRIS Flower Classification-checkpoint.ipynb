<h2> Methodology for Analyzing Features</h2>

The methodology for analyzing features in the Iris Flower Classification project adopts a structured approach to gain insights into the data and understand the relationships between measurements that distinguish each species.

* **Conduct an Overview of Data:**

    - Examine the dataset for missing values, duplicates, and overall data integrity.
    - Assess the distribution of each feature, including sepal and petal lengths and widths, and identify potential outliers.
    - Summarize the central tendency and variability of numerical variables to understand each species' characteristics.

* **Categorize Features:**

    - Organize features based on their relevance to the species classification task, specifically focusing on petal and sepal measurements.
    - Identify key measurements likely to contribute to species differentiation, based on prior biological knowledge.

* **Formulate Hypotheses:**

    - Define the null hypothesis (H0) which assumes that there are no significant differences between species based on sepal and petal measurements. This will be tested to confirm the usefulness of each feature in classification.

* **Statistical Testing:**

    Use statistical methods to analyze relationships within the data:
    
    - **Analysis of Variance (ANOVA)** to compare measurements across species and test feature relevance for classification.
    - **Correlation Analysis** to assess relationships between measurements and understand interactions between petal and sepal dimensions.

* **Visual Analysis:**

    - Utilize visual tools such as scatter plots, pair plots, and heatmaps to examine separability between species and detect patterns.
    - Employ box plots, violin plots, and histograms to observe the distribution and variation of features across species, and to identify any species-specific tendencies.

* **Incorporate Observations:**

    - Based on statistical and visual analysis, document observations regarding how each feature (e.g., petal length, petal width) contributes to species classification.
    - Use these insights to inform the selection of features for the predictive model, enhancing model accuracy by focusing on the most distinguishing measurements.

```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt 
import seaborn as sns
import missingno as msno

from pandas.plotting import andrews_curves
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.svm import SVC
from sklearn.metrics import confusion_matrix,classification_report, accuracy_score
from sklearn.ensemble import RandomForestClassifier
plt.style.use("ggplot")

import warnings
warnings.filterwarnings("ignore")
```

```python
# Load the datset
iris_df=pd.read_csv('IRIS.csv')
iris_df.head()
```

# <div style="padding:20px;color:white;margin:0;font-size:35px;text-align:center;display:fill;border-radius:10px;background-color:#c54E58;overflow:hidden; font-family: 'Lucida Console'"><b> STATISTICAL ANALYSIS</b></div>
<a id="analysis"></a>

```python
# Set plot style
sns.set(style="whitegrid")

# Plot histograms for each feature
fig, axes = plt.subplots(2, 2, figsize=(12, 8))
fig.suptitle("Distribution of Sepal and Petal Features", fontsize=16)

# Plot each feature in a separate subplot
sns.histplot(iris_df['sepal_length'], bins=15, kde=True, color="skyblue", ax=axes[0, 0]).set(title="Sepal Length")
sns.histplot(iris_df['sepal_width'], bins=15, kde=True, color="salmon", ax=axes[0, 1]).set(title="Sepal Width")
sns.histplot(iris_df['petal_length'], bins=15, kde=True, color="lime", ax=axes[1, 0]).set(title="Petal Length")
sns.histplot(iris_df['petal_width'], bins=15, kde=True, color="orange", ax=axes[1, 1]).set(title="Petal Width")

plt.tight_layout(rect=[0, 0, 1, 0.96])
plt.show()


```

```python
# Check for any non-numeric or unusual values in numerical columns
iris_df[['sepal_length', 'sepal_width', 'petal_length', 'petal_width']].applymap(type).nunique()
```

```python
# Convert species to categorical if needed
iris_df['species'] = iris_df['species'].astype('category')

# Pair plot
sns.pairplot(iris_df, hue="species", markers=["o", "s", "D"], palette="Set2")
plt.suptitle("Pair Plot of Iris Features by Species", y=1.02)
plt.show()
```

```python
# Box plots to compare the distributions of each feature by species
fig, axes = plt.subplots(2, 2, figsize=(12, 10))
fig.suptitle("Feature Distributions by Species", fontsize=16)

# Create box plots for each feature by species
sns.boxplot(data=iris_df, x="species", y="sepal_length", palette="Set2", ax=axes[0, 0]).set(title="Sepal Length")
sns.boxplot(data=iris_df, x="species", y="sepal_width", palette="Set2", ax=axes[0, 1]).set(title="Sepal Width")
sns.boxplot(data=iris_df, x="species", y="petal_length", palette="Set2", ax=axes[1, 0]).set(title="Petal Length")
sns.boxplot(data=iris_df, x="species", y="petal_width", palette="Set2", ax=axes[1, 1]).set(title="Petal Width")

plt.tight_layout(rect=[0, 0, 1, 0.96])
plt.show()
```

```python
# Violin plots for each feature by species
fig, axes = plt.subplots(2, 2, figsize=(12, 10))
fig.suptitle("Violin Plots of Features by Species", fontsize=16)

sns.violinplot(data=iris_df, x="species", y="sepal_length", palette="Set2", ax=axes[0, 0]).set(title="Sepal Length")
sns.violinplot(data=iris_df, x="species", y="sepal_width", palette="Set2", ax=axes[0, 1]).set(title="Sepal Width")
sns.violinplot(data=iris_df, x="species", y="petal_length", palette="Set2", ax=axes[1, 0]).set(title="Petal Length")
sns.violinplot(data=iris_df, x="species", y="petal_width", palette="Set2", ax=axes[1, 1]).set(title="Petal Width")

plt.tight_layout(rect=[0, 0, 1, 0.96])
plt.show()
```

```python
# Swarm plots for each feature by species
fig, axes = plt.subplots(2, 2, figsize=(12, 10))
fig.suptitle("Swarm Plots of Features by Species", fontsize=16)

sns.swarmplot(data=iris_df, x="species", y="sepal_length", palette="Set2", ax=axes[0, 0]).set(title="Sepal Length")
sns.swarmplot(data=iris_df, x="species", y="sepal_width", palette="Set2", ax=axes[0, 1]).set(title="Sepal Width")
sns.swarmplot(data=iris_df, x="species", y="petal_length", palette="Set2", ax=axes[1, 0]).set(title="Petal Length")
sns.swarmplot(data=iris_df, x="species", y="petal_width", palette="Set2", ax=axes[1, 1]).set(title="Petal Width")

plt.tight_layout(rect=[0, 0, 1, 0.96])
plt.show()
```

```python
# Andrews curves plot
plt.figure(figsize=(10, 6))
andrews_curves(iris_df, "species", colormap="Set1")
plt.title("Andrews Curves for Iris Species")
plt.show()
```

# <div style="padding:20px;color:white;margin:0;font-size:35px;text-align:center;display:fill;border-radius:10px;background-color:#c54E58;overflow:hidden; font-family: 'Lucida Console'"><b>MODEL TRAINING AND EVALUATION</b></div>
<a id="model"></a>

```python
# Prepare features and target variable
X = iris_df.drop(columns="species")
y = iris_df["species"]
```

```python
# Split the dataset
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

# <div style="padding:15px;color:white;margin:0;font-size:25px;text-align:center;display:fill;border-radius:5px;background-color:blue;overflow:hidden; font-family: 'Lucida Console'"><b> 1. RANDOM FOREST CLASSIFIER</b></div>
<a id="rf"></a>

```python
# Initialize and train the model
rf = RandomForestClassifier(random_state=42)
rf.fit(X_train, y_train)

# Predictions
y_pred = rf.predict(X_test)

# Evaluation
print("Confusion Matrix:\n", confusion_matrix(y_test, y_pred))
print("\nClassification Report:\n", classification_report(y_test, y_pred))
```

# <div style="padding:15px;color:white;margin:0;font-size:25px;text-align:center;display:fill;border-radius:5px;background-color:blue;overflow:hidden; font-family: 'Lucida Console'"><b>2.  SUPPORT VECTOR MACHINE (SVM)</b></div>
<a id="gb"></a>

```python
# Initialize and train the SVM model
svm_model = SVC(kernel='linear', C=1, random_state=42)
svm_model.fit(X_train, y_train)

# Evaluate the SVM model with cross-validation
svm_cv_scores = cross_val_score(svm_model, X, y, cv=5)
print("SVM Cross-validation scores:", svm_cv_scores)
print("SVM Mean CV accuracy:", svm_cv_scores.mean())

# Predict and evaluate on the test set
y_pred_svm = svm_model.predict(X_test)
print("SVM Test Set Accuracy:", accuracy_score(y_test, y_pred_svm))
print("\nSVM Classification Report:\n", classification_report(y_test, y_pred_svm))
print("\nSVM Confusion Matrix:\n", confusion_matrix(y_test, y_pred_svm))
```

```python
# Plot feature importance for Random Forest
importances = rf.feature_importances_
features = X.columns
importance_df = pd.DataFrame({"Feature": features, "Importance": importances}).sort_values(by="Importance", ascending=False)

plt.figure(figsize=(10, 6))
sns.barplot(x="Importance", y="Feature", data=importance_df)  # Corrected line
plt.title("Feature Importance - Random Forest")
plt.show()
```
